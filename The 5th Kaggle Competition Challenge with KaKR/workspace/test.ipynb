```python
from sklearn import metrics
import numpy as np

SUBGROUP_AUC = 'subgroup_auc'
BPSN_AUC = 'bpsn_auc'  # stands for background positive, subgroup negative
BNSP_AUC = 'bnsp_auc'  # stands for background negative, subgroup positive

def compute_auc(y_true_proba, y_pred_proba):
    try:
        return metrics.roc_auc_score(y_true_proba >= 0.5, y_pred_proba >= 0.5)
    except ValueError:
        return np.nan

def compute_subgroup_auc(df, subgroup_col, true_col, pred_col):
    subgroup_indices = df[subgroup_col] >= 0.5
    subgroup_examples = df[subgroup_indices]
    
    return compute_auc(subgroup_examples[true_col], subgroup_examples[pred_col])

def compute_bpsn_auc(df, subgroup_col, true_col, pred_col):
    """Computes the AUC of the within-subgroup negative examples and the background positive examples."""
    subgroup_indices = df[subgroup_col] >= 0.5
    label_indices = df[true_col] >= 0.5
    
    subgroup_negative_examples = df[subgroup_indices & ~label_indices]
    non_subgroup_positive_examples = df[~subgroup_indices & label_indices]
    examples = pd.concat([subgroup_negative_examples, non_subgroup_positive_examples], axis=0)
    
    return compute_auc(examples[true_col], examples[pred_col])

def compute_bnsp_auc(df, subgroup_col, true_col, pred_col):
    """Computes the AUC of the within-subgroup positive examples and the background negative examples."""
    subgroup_indices = df[subgroup_col] >= 0.5
    label_indices = df[true_col] >= 0.5
                                        
    subgroup_positive_examples = df[subgroup_indices & label_indices]
    non_subgroup_negative_examples = df[~subgroup_indices & ~label_indices]
    examples = pd.concat([subgroup_positive_examples, non_subgroup_negative_examples], axis=0)
    return compute_auc(examples[true_col], examples[pred_col])

def compute_bias_metrics_for_model(dataset, subgroups, true_col = "toxicity", pred_col = "predict"):
    """Computes per-subgroup metrics for all subgroups and one model."""
    records = []
    for subgroup in subgroups:
        record = {
            'subgroup': subgroup,
            'subgroup_size': (dataset[subgroup] >= 0.5).sum()
        }
        record[SUBGROUP_AUC] = compute_subgroup_auc(dataset, subgroup, true_col, pred_col)
        record[BPSN_AUC] = compute_bpsn_auc(dataset, subgroup, true_col, pred_col)
        record[BNSP_AUC] = compute_bnsp_auc(dataset, subgroup, true_col, pred_col)
        records.append(record)

    return pd.DataFrame(records).sort_values('subgroup_auc', ascending=True)

# bias_metrics_df = compute_bias_metrics_for_model(val_dataset, IDENTITY_COLMUNS)
# bias_metrics_df
```
